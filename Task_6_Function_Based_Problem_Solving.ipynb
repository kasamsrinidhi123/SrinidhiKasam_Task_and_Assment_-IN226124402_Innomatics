**Problem Statement 1:Smart Parking Lot Management system**

```python
def smart_parking_system(capacity, vehicle_logs):
    current_parked = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            if current_parked > 0:
                current_parked -= 1

        # track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked

    # parking status check
    if current_parked > capacity:
        status = "ALERT: Parking Full / Exceeded Capacity"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage)
    print("Parking Status:", status)
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)
```

**Problem Statement 2:Online Food Delivery Time Estimator**

```python
def estimate_delivery_time(distance, traffic, weather):
    # Base delivery time (minutes per km)
    base_time_per_km = 5
    total_time = distance * base_time_per_km

    # Traffic delay (minutes)
    traffic = traffic.lower()
    if traffic == "low":
        traffic_delay = 0
    elif traffic == "medium":
        traffic_delay = 5
    elif traffic == "high":
        traffic_delay = 10
    else:
        traffic_delay = 0

    # Weather delay (minutes)
    weather = weather.lower()
    if weather == "clear":
        weather_delay = 0
    elif weather == "cloudy":
        weather_delay = 3
    elif weather == "rainy":
        weather_delay = 5
    elif weather == "stormy":
        weather_delay = 8
    else:
        weather_delay = 0

    total_time += traffic_delay + weather_delay
    return total_time

eta = estimate_delivery_time(distance, traffic, weather)

print(f"\nEstimated Delivery Time: {int(eta)} minutes")
```

**Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**

```python
def analyze_seat_occupancy(total_seats, booked_seats):
    # count booked seats
    booked_count = sum(booked_seats)

    # calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    # suggest additional show
    if occupancy >= 90:
        suggestion = "Open additional show"
    else:
        suggestion = "No additional show needed"

    # display results
    print(f"Occupancy: {int(occupancy)}%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)


# ---- Sample Input ----
total_seats = 200
booked_seats = [1] * 150   # 150 booked seats

# ---- Function Call ----
analyze_seat_occupancy(total_seats, booked_seats)
```

 **Problem Statement 4:Cloud Server Load Classification System**

```python
def classify_server_load(cpu_readings):
    # calculate average CPU usage
    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    # classify server load
    if avg_cpu < 50:
        status = "Normal"
    elif avg_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    # display results
    print(f"Average CPU Load: {int(avg_cpu)}%")
    print("Server Status:", status)


# ---- Sample Input ----
cpu_readings = [45, 60, 70, 85, 90]

# ---- Function Call ----
classify_server_load(cpu_readings)
```

 **Problem Statement 5:Smart Classroom Resource Usage Monitor**

```python
def monitor_resource_usage(resource_usage):
    overused = []

    # define overuse threshold (hours)
    threshold = 8

    # check each resource
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)

    # display results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# ---- Sample Input ----
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# ---- Function Call ----
monitor_resource_usage(usage)
```

**Problem Statement 6: Online Event Registration Capacity Controller**

```python
def manage_registrations(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0

    # simulate registration process
    for _ in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # determine registration status
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"

    # display results
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# ---- Sample Input ----
event_capacity = 100
registrations = 105

# ---- Function Call ----
manage_registrations(event_capacity, registrations)
```
