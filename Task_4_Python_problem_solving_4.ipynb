**Problem 1: Mobile Recharge Validation System**

```python
def validate_recharge(amount,valid_plans):
  if amount<50:
    print("recharge amount must be 50")
  if amount not in valid_plans:
    print("plan not avialable")
    return False
  return True
valid_plans = [199, 299, 399, 599]
while True:
  amount = int(input("enter recharge amount: "))
  if validate_recharge(amount,valid_plans):
    print("Recharge successful")
  elif amount==0:
    break
  else:
    print("Recharge failed")
```

**Problem 2: Inventory Reorder Alert System**

```python
def check_stock(stock):
  for product, quantity in stock.items():
    if quantity<15:
       print(product,"-Recorder Alert")
    else:
       print(product,"-Stock OK")
stock={
    "rice":15,
    "oil":10,
    "milk":23,
    "choclate":7
}
check_stock(stock)
```

**Problem 3: Student Result Processing System**

```python
def check_marks(marks):
  total=0
  #calculate total
  for m in marks:
    total+=m
  average=total/len(marks)
  if average>=50:
    print("average:",average)
    return "Pass"
  else:
    return "Fail"
marks=[30,50,70,60]
check_marks(marks)
```

**Problem 4: Cab Fare Estimator with Retry Option**

```python
def calculate_fare(distance,peak):
  Base_fare = 50
  km_rate=12
  fare=Base_fare+(distance*km_rate)
  #add peak hr charge
  if peak.lower()=="yes":
    fare+=fare*0.25
  return fare
while True:
  distance = float(input("Enter distance travelled (km): "))
  peak = input("Is it peak hour? (yes/no): ")

  total_fare = calculate_fare(distance, peak)
  print("Total Fare = ₹", total_fare)

  retry = input("Do you want to calculate again? (yes/no): ")
  if retry.lower() != "yes":
      break
```

**Problem 5: Employee Attendance Eligibility Checker**

```python
def check_attendance(attendance):
    present= 0
    # count present days
    for day in attendance:
        if day == "P":
            present+= 1
    total_days = len(attendance)
    percentage = (present / total_days) * 100
    print(f"\nAttendance Percentage: {percentage}%")
    # check eligibility
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
# example of attendance list
attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]
result = check_attendance(attendance_list)
print("Attendance Result:", result)
```

**Problem 6: Password Strength Checker**

```python
def check_password(password):
  has_digit = False
  has_special = False
  special_chars = "@#$"
  # check characters using loop
  for ch in password:
      if ch.isdigit():
          has_digit = True
      if ch in special_chars:
          has_special = True
  # check all conditions
  if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
  else:
        return "Weak Password"
# example
check= input("Enter password: ")
result = check_password(check)
print(result)
```
