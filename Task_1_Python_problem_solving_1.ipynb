```python
#1. User Login Check
def login_check():
  username="admin"
  password="1234"
  user_name=input("Enter username: ")
  pass_word=input("Enter password: ")
  if user_name==username and pass_word==password:
    print("Login Successful")
  else:
    print("Invalid Credentials")

if __name__=="__main__":
  login_check()
```

```python
#2.Pass / Fail Analyzer
def student_results():
  marks=[45,78,90,33,60]
  passmark=0
  failmark=0
  for mark in marks:
    if mark>=50:
      passmark+=1
    else:
      failmark+=1
  print("###Student results###")
  print("total students:",len(marks))
  print("No.of students Passed: ",passmark)
  print("No.of students Failed: ",failmark)
if __name__=="__main__":
  student_results()
```

```python
#3. Simple Data Cleaner
def clean_data():
  user_names=[" Alice ", "bob", " CHARLIE "]
  updated_list=[name.strip().lower() for name in user_names]
  print("Cleaned names list:")
  print(updated_list)
if __name__=="__main__":
  clean_data()
```

```python
#4. Message Length Analyzer
def check_message():
  messages = ["Hi", "Welcome to the platform", "OK"]
  for message in messages:
    length = len(message)
    status="long message " if length >10 else "short message"
    print(f"{message} is {status}")
if __name__=="__main__":
  check_message()
```

```python
#5.Error Message Detector
def check_error_message():
  logs = ["INFO", "ERROR", "WARNING", "ERROR"]
  total_errors=logs.count("ERROR")
  print(f"Total errors: {total_errors}")
if __name__=="__main__":
  check_error_message()
```
