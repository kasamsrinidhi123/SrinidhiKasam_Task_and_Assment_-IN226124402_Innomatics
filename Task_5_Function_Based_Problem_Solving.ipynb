**Problem 1: Social Media Post Engagement Analyzer**

```python
def post_analyzer(likes_list):
    total_likes = 0
    for likes in likes_list:
        total_likes += likes
    print("Total Likes:", total_likes)
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")
data = [200, 300, 150, 400]
post_analyzer(data)
```

**Problem 3: Agriculture – Rainfall Adequacy Checker**

```python
def rainfall_checker(rainfall_data, required_level):
    total = 0
    #  to calculate total rainfall
    for rain in rainfall_data:
        total += rain
    # calculate average rainfall
    average = total / len(rainfall_data)
    print("Average Rainfall:", int(average))

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
# Example
season_data = [70, 75, 68, 80, 67]
required = 70
rainfall_checker(season_data, required)
```

**Problem 4: Social Media – Duplicate Account Detection**

```python
def duplicate_account(usernames):
  unique_usernames=set(usernames)
  #loop to find
  if len(unique_usernames)<len(usernames):
    print("Duplicate Account Found: Yes")
  else:
    print("Duplicate Accounts Found: No")
  #example
accounts=["sri123","sita456","ram123","ram123"]
duplicate_account(accounts)
```

**Problem 6: Agriculture – Premium Crop Price Filter**

```python
def premium_crop(prices):
    premium = []
    # premium prices
    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)
crop_prices = [1500, 2500, 1800, 3200, 900]
premium_crop(crop_prices)
```

**Problem 7: System Monitoring – Application Health Checker**

```python
def application_health(errors):
    print("Error Count:", errors)
    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")
application_health(7)
```

**Problem 9: E-Learning – Student Attendance Eligibility System**

```python
def attendance_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended = 0
    for status in attendance_list:
        if status== 1:
            attended += 1
    percentage = (attended / total_classes) * 100
    print("Attendance Percentage:", percentage)
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
student_attendance = [1, 1, 0, 1, 1]
attendance_eligibility(student_attendance)
```

**Problem 10: Smart Electricity Bill Analyzer**

```python
def smart_electricity(units):
    bill = 0
    # Calculate bill
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    # Classify usage
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    return bill, usage
#example
units_consumed = 250
total_bill, usage_status = smart_electricity(units_consumed)
print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
